```python
import numpy as np
import random as rn
import matplotlib.pyplot as plt
import sys
n=12

D = [[0 for _ in range(n)] for _ in range(n)]
for i in range(n):
    D[i][i] = i+1
D = np.array(D)
rn.seed(1000)
Q = np.array([[rn.randint(0,6) for _ in range(n)] for _ in range(n)])
Q = np.linalg.qr(Q)[0]
Q1 = np.linalg.inv(Q)
A = np.dot(Q1, D)
A = np.dot(A, Q)
with open('Ort.txt', 'w') as file:
    for row in Q: 
        print(*row, file=file)
        
with open('Ort_T.txt', 'w') as file:
    for row in Q1: 
        print(*row, file=file) 
    
x_i = [i*(-1)**i for i in range(1,13)]
print(np.dot(A,x_i))
```
