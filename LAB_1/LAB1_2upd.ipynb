```python
import matplotlib.pyplot as plt
mpd = [[], [], [], [], []]
with open('MPD.TXT', mode='rt') as file:
    for line in file:
        line = line.split()
        for i in range(5):
            mpd[i].append(float(line[i]))

mmn = [[], [], [], [], []]
with open('MMN.TXT', mode='rt') as file:
    for line in file:
        line = line.split()
        for i in range(5):
            mmn[i].append(float(line[i]))            
```

```python
plt.loglog(mpd[0], mpd[1], color='red', linestyle='dashed')
plt.loglog(mpd[0], mpd[3], color='blue', linestyle='dashed')
plt.loglog(mmn[0], mmn[1], color='red')
plt.loglog(mmn[0], mmn[3], color='blue')
plt.title("Зависимость фактическо ощибки от точности")
plt.xlabel("eps", fontsize=14)          
plt.ylabel("|x*-x|", fontsize=14)
plt.legend(["МПД алг.", "МПД трансц.", "ММН алг.", "ММН трансц."])
plt.grid()
plt.show()
```

```python
plt.semilogx(mpd[0], mpd[2], color='red', linestyle='dashed')
plt.semilogx(mpd[0], mpd[4], color='blue', linestyle='dashed')
plt.semilogx(mmn[0], mmn[2], color='red')
plt.semilogx(mmn[0], mmn[4], color='blue')
plt.title("Зависимость числа итераций от точности")
plt.xlabel("eps", fontsize=14)          
plt.ylabel("i", fontsize=14)
plt.legend(["МПД алг.", "МПД трансц.", "ММН алг.", "ММН трансц."])
plt.grid()
plt.show()
```
