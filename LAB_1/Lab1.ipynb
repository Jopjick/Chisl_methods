```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f1(x):
    return 2*x**4 - 9*x**3 - 60*x**2 + 1
def f2(x):
    from math import log10
    return x*log10(x) - 1

args1 = np.linspace(-4, -3, 100)
args2 = np.linspace(1, 5, 500)
y1 = [f1(x) for x in args1]
y2 = [f2(x) for x in args2]
```

```python
plt.plot(args1, y1)
plt.title('Зависимость y = 2*x**4 - 9*x**3 - 60*x**2 + 1')
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)") 
plt.show()

plt.plot(args2, y2, color='orange')
plt.title('Зависимость y = x*log10(x) - 1')
plt.grid()
plt.xlabel("x")
plt.ylabel("g(x)") 
plt.show()
```
