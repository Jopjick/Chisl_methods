```python
import numpy as np
import random
```

```python
random.seed(1000)

U = [[0 for _ in range(12)] for _ in range(12)]
for i in range(12):
    for j in range(12):
        if i != j and i <= j:
            U[i][j] = random.randint(0,10)/100
        elif i==j:
            U[i][j] = 1          
U = np.array(U)
with open('A.txt', 'w') as file:
    for row in U:
        print(*row, file=file)

U[1][1]=1
random.seed(100000)

Q = np.array([[random.randint(0,6) for _ in range(12)] for _ in range(12)])
Q = np.linalg.qr(Q)[0]
Q1 = np.linalg.inv(Q)
A = np.dot(Q1, U)
A = np.dot(A, Q)

print(np.linalg.cond(U))
     
random.seed(10000)   

X = np.array([random.randint(-5,5) for _ in range(12)])

B = np.dot(X,A)

with open('Ort.txt', 'w') as file:
    for row in Q: 
        print(*row, file=file)
        
with open('Ort_T.txt', 'w') as file:
    for row in Q1: 
        print(*row, file=file)        
        
with open('X.txt', 'w') as file:
    print(*X, file=file)
    
with open('B.txt', 'w') as file:
    print(*B, file=file)
```
